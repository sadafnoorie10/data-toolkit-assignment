Q3.  Explain the following terms:
 The difference in np.array, np.asarray and np.asanyarray.
 The difference between Deep copy and shallow copy.

A3. ### **Difference Between `np.array`, `np.asarray`, and `np.asanyarray`:**

#### 1. **`np.array`:**
   - Always creates a **new NumPy array**.
   - If the input is already a NumPy array, it creates a **copy** by default.
   - You can control whether to make a copy using the `copy` parameter (`copy=True` by default).
   - **Example:**
     ```python
     import numpy as np
     data = [1, 2, 3]
     arr = np.array(data)  # Creates a new NumPy array
     ```

#### 2. **`np.asarray`:**
   - Converts the input to a NumPy array but **does not copy the data** if the input is already a NumPy array.
   - Faster than `np.array` when no copy is needed.
   - **Example:**
     ```python
     data = np.array([1, 2, 3])
     arr = np.asarray(data)  # No copy is created, reuses the same data
     ```

#### 3. **`np.asanyarray`:**
   - Similar to `np.asarray`, but it **preserves the subclass of the input array** (e.g., `matrix` or `masked array`).
   - Useful when working with custom array types that extend NumPy arrays.
   - **Example:**
     ```python
     data = np.matrix([[1, 2], [3, 4]])
     arr = np.asanyarray(data)  # Retains the matrix subclass
     ```

---

### **Key Differences:**

| Function        | Copies Data  | Preserves Subclass |
|-----------------|--------------|--------------------|
| `np.array`      | Yes (default) | No                 |
| `np.asarray`    | No (if input is array) | No        |
| `np.asanyarray` | No (if input is array) | Yes       |

---

### **Difference Between Deep Copy and Shallow Copy:**

#### 1. **Shallow Copy:**
   - Creates a **new object** but **references the same child elements** as the original.
   - Changes to mutable elements (e.g., lists) in the original object will be reflected in the copy.
   - **Example:**
     ```python
     import copy
     original = [[1, 2], [3, 4]]
     shallow = copy.copy(original)
     shallow[0][0] = 99  # Modifies the original
     print(original)  # [[99, 2], [3, 4]]
     ```

#### 2. **Deep Copy:**
   - Creates a **new object** and **recursively copies all elements**, making the copy completely independent of the original.
   - Changes to the original do not affect the copy.
   - **Example:**
     ```python
     deep = copy.deepcopy(original)
     deep[0][0] = 42  # Does not modify the original
     print(original)  # [[1, 2], [3, 4]]
     ```

---

### **Key Differences:**

| Feature                 | Shallow Copy                          | Deep Copy                              |
|-------------------------|----------------------------------------|----------------------------------------|
| Copies Parent Object    | Yes                                   | Yes                                    |
| Copies Child Objects    | No (references original child objects) | Yes (creates new child objects)        |
| Memory Usage            | Lower                                 | Higher                                 |
| Independence            | No (changes in one affect the other)  | Yes (completely independent objects)   |

Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

````python
A17.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays (x and y)
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame with x and y and assign quadrant labels
data = pd.DataFrame({'x': x, 'y': y})

# Define a function to classify points based on their position relative to the origin
def classify_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant I'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant II'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to classify the quadrants
data['Quadrant'] = data.apply(classify_quadrant, axis=1)

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='Quadrant', palette='coolwarm', data=data)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrant')

# Show the plot
plt.tight_layout()
plt.show()
````

Explanation:
Data Generation: Two random arrays x and y are created using np.random.randn(100) to generate 100 samples from a normal distribution (mean 0, standard deviation 1).
Quadrant Classification: A function classify_quadrant is defined to categorize each point into one of the four quadrants based on the signs of x and y.
DataFrame: The pandas.DataFrame is used to store the data and the quadrant classification.
Seaborn Scatter Plot: sns.scatterplot is used to create the scatter plot. The points are colored based on the 'Quadrant' column, with a color palette ('coolwarm').
Labels and Title: The x-axis and y-axis are labeled, and the title of the plot is set to 'Quadrant-wise Scatter Plot'.
Legend: The legend is displayed to show the quadrant labels.

````python
A19.
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np

# Enable output in the notebook
output_notebook()

# Random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=5)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create the figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="",
           x_axis_label='Category', y_axis_label='Value', height=350)

# Add the bars to the figure
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field="categories", line_color="white",
       fill_color='colors')

# Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Color the bars based on values
color_map = ['#FF6347', '#FFD700', '#32CD32', '#1E90FF', '#8A2BE2']
source.data['colors'] = color_map  # Assign colors to bars

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.location = "top_left"

# Display the plot
show(p, notebook_handle=True)
````

Explanation:
Output for Colab: output_notebook() is required in Colab to display the plot inline.
Random Data: The categories are predefined ('A', 'B', etc.), and random values between 1 and 20 are assigned to them.
ColumnDataSource: This is used to manage the data in Bokeh and update the plot dynamically.
Bar Chart: Bars are created using p.vbar() with the color assigned to them based on the values.
HoverTool: The hover tool is added to display the category and value when hovering over the bars.
Color Map: A color map (color_map) is used to color the bars based on their values.
